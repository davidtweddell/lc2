```python
def splitCategory(x):
    if '-' in x:
        return x.split('-')[0].capitalize()
    return 'Demo'
```

```python
lCategory = [splitCategory(x) for x in lCols]
```

```python
dfClean = pd.DataFrame(np.vstack([lCols,lCategory]).T,columns=['feature','category'])
```

```python
def cleanCols(x):
    cat = x['category']
    feat = x['feature'].replace('shortness_of_breath_dyspne','Dyspnea').replace('runny_nose_rhinorrhea','Rhinorrhea').replace('muscle_aches_myalgia','Myalgia').replace('loss_of_taste_lost_of_smel', 'Loss of Taste/Smell').replace('joint_pain_arthralgia','Arthralgia').replace('confusion_altered_mental_s','Altered Mental Status')
    
    if 'fever_38_0oc___1' in feat.lower(): return '(Sym_Cov) Fever > 38'
    if 'fever_38_0oc___2' in feat.lower(): return '(Sym_LC) Fever > 38'

    if feat.lower() == 'subst-3':
        return 'Alcohol Intake'
    
    if 'vaccinated' in feat: return 'Vaccinated'
    
    if cat == 'Demo':
        return feat.capitalize().strip()
    if cat =='Med': 
        featMain = feat.replace('-agents','').replace('MED-','').replace('-',' ').replace('Anti','Anti-').title()
        return ('(Hx_Med) '+featMain).strip()
    if cat =='Sympt':
        featMain = feat.replace('SYMPT-','').replace('-',' ').replace('_',' ')
        if feat[-1]=='1':
            return ('(Cov_Sym) '+featMain.replace('1', '').title()).strip()
        if feat[-1]=='2':
            return  ('(LCov_Sym) ' + featMain.replace('2','').title()).strip()
    if cat =='Hhx':
        featMain = feat.replace('HHX-OTHER-','').replace('HHX-','').replace('-',' ').title()
        return ('(Hx) '+featMain).strip()
    
    if cat =='Cd':
        medMap = {
            '1': 'Steroids',
            '2': 'Remdesivir',
            '3': 'Anticoagulant',
            '4': 'Colchicine',
            '5': 'Antibiotics',
            '6': 'Other',
        }
        if 'medication_received' in feat:
            c = feat[-1]
            return '(Cov_Med) '+medMap[c]
        
        if 'severity' in feat:
            if feat[-1]=='0': return ('(Cov_Sevr) UnInfected')
            if feat[-1]=='1': return ('(Cov_Sevr) Mild')
            if feat[-1]=='2': return ('(Cov_Sevr) Moderate')
            if feat[-1]=='3': return ('(Cov_Sevr) Severe')
            if feat[-1]=='4': return ('(Cov_Sevr) Dead')
            
        
        if 'type_of_care___1' in feat: return '(Cov) ICU'
        if 'type_of_care___2' in feat: return '(Cov) Non-ICU'
        
        if 'memory_problems___4' in feat: return '(LCov_Comp) Memory Unknown Origin'
        if 'difficulty_with_concentrat___4' in feat: return '(LCov_Comp) Concentration Unknown Origin'
        if 'adjunctive___1' in feat: return '(Cov_Ind) Vasopressor/Inotrope'
        
        if '___' not in feat:
            featMain = feat.replace('CD-','')
            
            if featMain == 'required_oxygen': return '(Cov_Ind) Oxygen'
            if featMain == 'respiratory-other-4': return '(LCov_Comp) Respiratory Unknown Origin'
            
            
            if featMain == 'spo2': return '(Cov_Ind) SpO2'
            # if featMain == 'respiratory-other-4': efat
            return ('(Cov) '+featMain.title())
    
```

```python
dfClean['clean'] = dfClean.apply(cleanCols,axis=1)
def flipParen(x):
    if '('  in x:
        x = x.split(' ',1)
        return f"{x[1]} {x[0]}"
    return x
dfClean['clean_flip'] = dfClean['clean'].apply(flipParen)
```

```python
def make_categorical_colours(df, palette):

    # make a list of colours
    colours_dict = {feature: palette[i] for i, feature in enumerate(df)}
    return colours_dict
```
